```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df = pd.read_csv('cars.csv', skipinitialspace=True)
df
```

```python
'''
1. Create scatter plots for the following data, make sure all the plots appear in the same
plane with x-labels and y-labels evenly spaced.

  a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
  hp-time-to-60) with respect to each brand.

  mpg vs hp
  mpg vs weight
  hp vs weight
  hp vs time-to-60

'''
fig = plt.subplots(2,2 , figsize=(20,10))
plt.subplot(2,2,1)
sns.scatterplot(x = df["mpg"], y = df["hp"], data = df, hue = df["brand"])

plt.subplot(2,2,2)
sns.scatterplot(x = df["mpg"], y = df["weightlbs"], data = df, hue = df["brand"])

plt.subplot(2,2,3)
sns.scatterplot(x = df["hp"], y = df["weightlbs"], data = df, hue = df["brand"])

plt.subplot(2,2,4)
sns.scatterplot(x = df["hp"], y = df["time-to-60"], data = df, hue = df["brand"])

plt.show()
```

```python
'''2. Create bar plots for the following, make sure all the plots appear in the same plane with
x-labels and y-labels evenly spaced.

  a. Create a bar plot that shows the visual representation of hp, mpg, weight,
  time-to-60 with respect to the number of cylinders for each of the brands.

      hp vs cylinders
      mpg vs cylinders
      weight vs cylinders
      time-to-60 vs cylinders

'''
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
sns.barplot(x=df["cylinders"],   y=df["hp"],  data=df, hue=df["brand"])

plt.subplot(2,2,2)
sns.barplot(x=df["cylinders"],   y=df["mpg"],  data=df,  hue=df["brand"])

plt.subplot(2,2,3)
sns.barplot(x=df["cylinders"],   y=df["weightlbs"],  data=df,  hue=df["brand"])

plt.subplot(2,2,4)
sns.barplot(x=df["cylinders"],   y=df["time-to-60"],  data=df,  hue=df["brand"])
plt.show()

```

```python
''''
 b. Create a bar plot that shows the visual representation of hp, mpg, weight and
    time-to-60 with respect to the years, for each brand.

    HP vs year
    mpg vs year
    weight vs year
    timeto60 vs year

'''

fig= plt.subplots( 2,2, figsize=(30,20))

plt.subplot(2,2,1)
sns.barplot(x=df["year"], y=df["hp"] ,data=df , hue=df["brand"])

plt.subplot(2,2,2)
sns.barplot(x=df["year"], y=df["mpg"] ,data=df , hue=df["brand"])

plt.subplot(2,2,3)
sns.barplot(x=df["year"], y=df["weightlbs"] ,data=df , hue=df["brand"])

plt.subplot(2,2,4)
sns.barplot(x=df["year"], y=df["time-to-60"] ,data=df , hue=df["brand"])
plt.show()
```

```python
'''
3. Create box plots for the following:

  a. Create box plots for the columns in the step1 with respect to the number of
     cylinders and brand respectively.
     mpg vs cylinders vs brand
     hp vs cylinders vs brand
     weight vs cylinders vs brand
     time-to-60 vs cylinders vs brand
'''
fig=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.boxplot(x=df["cylinders"], y=df["mpg"], data=df, hue=df["brand"])

plt.subplot(2,2,2)
sns.boxplot(x=df["cylinders"], y=df["hp"], data=df, hue=df["brand"])

plt.subplot(2,2,3)
sns.boxplot(x=df["cylinders"], y=df["weightlbs"], data=df, hue=df["brand"])

plt.subplot(2,2,4)
sns.boxplot(x=df["cylinders"], y=df["time-to-60"], data=df, hue=df["brand"])

plt.show()

```

```python
'''
4. Create pair plots for the entire data to study various patterns in the data.
    a. Create pair plots with respect to brand, number of cylinders, year, etc.

'''
sns.pairplot(df, hue = "brand")
plt.show()
```

```python
'''
5. Create a heatmap for the entire data to study correlation between each of the columns.

'''
df_encoded = pd.get_dummies(df, columns=['brand'], drop_first=True)
sns.heatmap(df_encoded.corr(), annot=True)
plt.show()
```

```python
'''
6. Create a histogram for the following:

  a. Create histograms for hp, mpg, time-to-60, weightlbs, cubic inches, year, with
     respect to brand.

     hp vs brand
     mpg vs brand
     cylinders vs brand
     time-to-60 vs brand
     weight lbs vs brand
     cubicinches vs brand
     year vs brand
'''
fig=plt.subplots(2,3,figsize=(20,10))

plt.subplot(2,3,1)
sns.histplot(x=df["hp"], data=df, hue=df["brand"])

plt.subplot(2,3,2)
sns.histplot(x=df["mpg"], data=df, hue=df["brand"])

plt.subplot(2,3,3)
sns.histplot(x=df["time-to-60"], data=df, hue=df["brand"])

plt.subplot(2,3,4)
sns.histplot(x=df["weightlbs"], data=df, hue=df["brand"])

plt.subplot(2,3,5)
sns.histplot(x=df["cubicinches"], data=df, hue=df["brand"])

plt.subplot(2,3,6)
sns.histplot(x=df["year"], data=df, hue=df["brand"])
plt.show()

```

```python
'''
7. Create line plots for the following:

    a. Line plot between hp-mpg, hp-(time-to-60), mpg-(time-to-60), hp-weight lbs,
    mpg-weight lbs, cubic inches-weight lbs.

    hp vs mpg
    hp vs time-to-60
    mpg vs time-to-60
    hp vs weight lbs
    mpg vs weight lbs
    cubic inches vs weight lbs
'''
fig = plt.subplots(2,3,figsize=(20,10))

plt.subplot(2,3,1)
sns.lineplot(x = df["hp"], y = df["mpg"], data = df, hue = df["brand"])

plt.subplot(2,3,2)
sns.lineplot(x = df["hp"], y = df["time-to-60"], data = df, hue = df["brand"])

plt.subplot(2,3,3)
sns.lineplot(x = df["mpg"], y = df["time-to-60"], data = df, hue = df["brand"])

plt.subplot(2,3,4)
sns.lineplot(x = df["hp"], y = df["weightlbs"], data = df, hue = df["brand"])

plt.subplot(2,3,5)
sns.lineplot(x = df["mpg"], y = df["weightlbs"], data = df, hue = df["brand"])

plt.subplot(2,3,6)
sns.lineplot(x = df["cubicinches"], y = df["weightlbs"], data = df, hue = df["brand"])

plt.show()
```

```python
'''
8. Create a pie chart for the following columns and their distribution in the data:

   a. Brands
'''

data = df['brand'].value_counts()
plt.pie(data, labels=data.index, autopct='%0.0f%%')
plt.title('Distribution of Brands')
plt.show()
```

```python
'''
9. Create violin plots for the following:

   a. hp, mpg, time-to-60, with respect to the number of cylinders for each brand.

  hp vs cylinders
  mpg vs cylinders
  time-to-60 vs cylinders

'''

fig = plt.subplots(1,3, figsize=(15,5))

plt.subplot(1,3,1)
sns.violinplot(x=df["cylinders"], y=df["hp"], data=df, hue=df["brand"])

plt.subplot(1,3,2)
sns.violinplot(x=df["cylinders"], y=df["mpg"], data=df, hue=df["brand"])

plt.subplot(1,3,3)
sns.violinplot(x=df["cylinders"], y=df["time-to-60"], data=df, hue=df["brand"])

plt.show()
```

```python
'''
10. Create categorical plots for the following:

    a. Hp, mpg, time-to-60, weight, cubic inches, year with respect to each brand for various
       numbers of cylinders.

       hp vs brand vs cylinders
       mpg vs brand vs cylinders
       time-to-60 vs brand vs cylinders
       weight vs brand vs cylinders
       cubicinches vs brand vs cylinders
       year vs brand vs cylinders

'''
fig = plt.subplots(5,1, figsize=(10,20))

#Bar-Plot
plt.subplot(5,1,1)
sns.barplot(x=df["cylinders"], y=df["hp"], data=df, hue=df["brand"])

#Violin-Plot
plt.subplot(5,1,2)
sns.violinplot(x=df["cylinders"], y=df["mpg"], data=df, hue=df["brand"])
plt.tight_layout()

#Line-Plot
plt.subplot(5,1,3)
sns.lineplot(x=df["cylinders"], y=df["time-to-60"], data=df, hue=df["brand"])
plt.tight_layout()

#Hist-Plot
plt.subplot(5,1,4)
sns.histplot(x=df["cylinders"], y=df["weightlbs"], data=df, hue=df["brand"])
plt.tight_layout()

#Box-Plot
plt.subplot(5,1,5)
sns.boxplot(x=df["cylinders"], y=df["year"], data=df, hue=df["brand"])
plt.tight_layout()
plt.show()
```

```python
'''
11. Create the area charts for the following:

    a. Mpg with respect to each brand of the car

    mpg vs brand
'''
fig = plt.subplots(1,1,figsize=(20,10))

plt.subplot(1,1,1)
sns.kdeplot(x="mpg", data=df, hue="brand", fill=True, alpha=0.5)
plt.show()
```

```python
'''
12. Create Doughnut plots to represent the following:

    a. The cars are based on various counts of cylinders

'''

# Calculate the distribution of cylinders
data = df['cylinders'].value_counts()

# Create the doughnut plot
plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.3})

# Add a title
plt.title('Distribution of Cars by Cylinder Counts')

# Show the plot
plt.show()
```
