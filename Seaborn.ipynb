```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal" ,data=fmri)
plt.show()
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = fmri)
plt.show()
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", marker = True,data = fmri)
plt.show()
```

```python
poke = pd.read_csv("pokemon.csv")
poke.head()
```

```python
sns.barplot(x="is_legendary", y="speed",data=poke)
plt.show()
```

```python
sns.barplot(x="is_legendary", y="weight_kg",data=poke, palette="rocket")
plt.show()
```

```python
sns.barplot(x="is_legendary", y="speed",data=poke, hue="generation")
plt.show()
```

```python
data = pd.read_csv("New_iris.csv")
data.head()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", data=data)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", data=data, hue="Species", style="Species")
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", data=data, hue="PetalLengthCm")
plt.show()
```

```python
dmd = pd.read_csv("diamonds.csv")
dmd.head()
```

```python
sns.distplot(dmd["price"])
plt.show()
```

```python
sns.kdeplot(dmd["price"])
plt.show()
```

```python
sns.kdeplot(dmd["price"],color = "g")
plt.show()
```

```python
sns.distplot(dmd["price"],color="purple",bins=10,kde=False)
plt.show()
```

```python
sns.distplot(dmd["price"],color="orange", vertical=True)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", data=data)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", data=data, color="brown")
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", data=data, color="olive",kind="reg")
plt.show()
```

```python
churn = pd.read_csv("churn.csv")
churn.columns
```

```python
sns.boxplot(x='Churn',y="tenure", data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y='MonthlyCharges',data=churn,palette="rocket")
plt.show()
```

```python
sns.boxplot(x="InternetService",y='tenure',data=churn,palette="Set1")
plt.show()
```

```python
sns.boxplot(x="Contract",y='tenure',data=churn,palette="Set1",linewidth=3)
plt.show()
```

```python
sns.boxplot(x="Contract",y='tenure',data=churn,palette="Set1",hue="PaymentMethod")
plt.show()
```
